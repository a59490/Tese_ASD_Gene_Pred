# Convert gene ids

```python
gene_symbols=file["gene_symbol"].copy()
gene_symbols=gene_symbols.dropna()
```

```python
from gprofiler import GProfiler

gp = GProfiler(return_dataframe=True)
gp_convert_df =gp.convert(organism  = 'hsapiens',
                            query= gene_symbols.tolist(),
                                target_namespace= 'ENSG',
                                numeric_namespace='SYMBOL',)
print(f'Total size: {len(gp_convert_df)}')
display(gp_convert_df["converted"].value_counts().head(10))
display(gp_convert_df.head())
```

```python
gp_convert_df=gp_convert_df.dropna() #---------------------------------------------------------------------------------------------------
gp_convert_df["converted"].value_counts().head(10)
```

### create a dict 

```python
convert_dict = dict(zip(gp_convert_df['incoming'], gp_convert_df['converted']))
```

## Merge the two

```python
def convert_gene_symbol_to_ensembl(gene_symbol):
    if gene_symbol not in convert_dict:
        return None
    else:
        ensenbl_id = convert_dict[gene_symbol]
        return ensenbl_id
```

```python
# gp_convert_df  file 

file["ensembl_id"]=file["gene_symbol"].apply(convert_gene_symbol_to_ensembl)
file
```

# Separate mutations

```python
file["inheritance"].unique()
```

```python
mutation_list=list(file["category"].unique())
inheritance_list_denovo=['d' ]
inheritance_list_inher=['i', 'm', 'p', 'b']


mutation_dict_denovo={}

mutation_dict_inher={}

grouped = file.groupby("category")

for mut, group_df in grouped:
    group_df= group_df.reset_index(drop=True).drop_duplicates(subset=["ensembl_id"])
    group_df = group_df[group_df["inheritance"]== "d"]
    mutation_dict_denovo[mut] = group_df


for mut, group_df in grouped:
    group_df= group_df.reset_index(drop=True).drop_duplicates(subset=["ensembl_id"])
    group_df = group_df[group_df["inheritance"].isin(inheritance_list_inher)]
    mutation_dict_inher[mut] = group_df
```

```python
mutation_dict_denovo["stopgain"]
```

```python
def make_gmt_file(mutation_dict, file_name):
    with open(file_name, "w") as f:
        for mut, df in mutation_dict.items():
            # Filter out None values from the gene list
            gene_list = [gene_id for gene_id in df["ensembl_id"] if gene_id is not None]
            gene_list = "\t".join(gene_list)
            f.write(f"{mut}\tMutation: {mut}\t{gene_list}\n")


make_gmt_file(mutation_dict_denovo, "mutation_denovo.gmt")
```

# Make The LOF 

add the framshift and stop mutations

```python
def merge_mutations(file_name,type):
    with open(file_name, "r") as f:
        lines = f.readlines()

    lof = "Lof\tMutation: Lof\t"
    lof_names = ["frameshift deletion", "frameshift insertion", "frameshift substitution", "stopgain", "stoploss"]
    syn_list = ["synonymous SNV","nonsynonymous SNV"]
    new_gmt = []

    for line in lines:
        line=line.strip("\n").split("\t")

        if line[0] in lof_names:
            genes=[gene for gene in line[2:] if gene != "None"]
            genes="\t".join(genes)
            lof=lof+genes
            print(lof)

        if line[0] in syn_list:
            line=[string for string in line if string != "None"]
            line="\t".join(line)
            new_gmt.append(line)
    new_gmt.append(lof)
    with open(f"merged_mutation_{type}.gmt", "w") as f:
        for line in new_gmt:
            f.write(f"{line}\n")

merge_mutations("mutation_denovo.gmt","denovo")
```
