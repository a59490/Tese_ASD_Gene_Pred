```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

emb=pd.read_csv('string_v12.csv')
emb["Unnamed: 0"]=emb["Unnamed: 0"].str.split('.').str[1]
prot_ids=pd.read_csv('prot_converter.csv')

print(f"emb {len(emb)}")
display(emb.head())

print(f"prot_ids {len(prot_ids)}")
display(prot_ids.head())
```

## Merge the two

```python
embedding_file = pd.merge(prot_ids, emb, left_on='ensb_prot_id', right_on='Unnamed: 0', how='inner')
embedding_file
```

```python
embedding_file.to_csv('string_embedding.csv')
```

## test the new embeddings

```python
import pandas as pd
gene_list=pd.read_csv('gene_lists/gene_list_krs_clean.csv')
gene_list
```

```python
positive_genes=gene_list[gene_list['y']==1]
negative_genes=gene_list[gene_list['y']==0]
```

```python
print(f"positive genes {len(positive_genes)}")
print(f"negative genes {len(negative_genes)}")
```

```python
positive_genes_l=set(positive_genes["gene"].to_list())
negative_genes_l=set(negative_genes["gene"].to_list())
emb_gene_set=set(embedding_file["ensb_gene_id"].to_list())
```

```python
print(f'number of positive genes in the emb file: {len(positive_genes_l & emb_gene_set)}')
print(f"number of negative genes in the emb file: {len(negative_genes_l & emb_gene_set)}")
print(f"number of negative + positives in the emb file: {len((positive_genes_l|negative_genes_l))}")
print(f"total number of genes from the gene list in the emb file: {len((positive_genes_l|negative_genes_l) & emb_gene_set)}")
print(f"total number of genes in the emb file: {len(emb_gene_set)}")
```
