```python
from grape import Graph
```

```python
# Define the file path and separator
csv_file_path = "9606.protein.links.v12.0.txt"

file_path = "9606.protein.links.v12.0.txt.gz"
separator = " "  # Assuming space is the separator, adjust if necessary

# Create the graph from the CSV file
string_graph = Graph.from_csv(
    edge_path=csv_file_path,
    edge_list_separator=separator,
    sources_column_number=0,  # Assuming protein 1 is in the first column
    destinations_column_number=1,  # Assuming protein 2 is in the second column
    weights_column_number=2,  # Assuming weight is in the third column ------------- /1000 check later!!!!!!!!
    directed=False,  # Assuming the graph is undirected, adjust if necessary
    verbose=True  # Optionally, set verbose to True for more information during loading
)
```

```python
string_graph
```

```python
string_graph.remove_disconnected_nodes()
```

```python
from grape import GraphVisualizer
from grape.embedders import FirstOrderLINEEnsmallen
```

```python
embedding = FirstOrderLINEEnsmallen().fit_transform(string_graph)
#embedding = Node2VecSkipGramEnsmallen().fit_transform(graph)
embedding
```

```python
embedding.to_csv("string_v12.csv")
```
