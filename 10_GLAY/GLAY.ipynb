```python
import igraph as ig
from gprofiler import GProfiler
import leidenalg as la
import pandas as pd
import matplotlib.pyplot as plt
import random
import matplotlib.pyplot as plt
import networkx as nx
from netgraph import Graph
import seaborn as sns
```

```python
def gprofiler(genes):
    gp = GProfiler(return_dataframe=True)
    gp_results = gp.profile(organism='hsapiens', query=genes)
    return gp_results
```

```python
glay= pd.read_csv('data_filtered.csv--clustered default  node.csv')
leinden = pd.read_csv('leiden.csv')

leinden = leinden[['name','__leidenCluster']]
glay=glay[['name','__glayCluster']]

glay=glay.rename(columns={'name':'name','__glayCluster':'cluster'})
leinden=leinden.rename(columns={'name':'name','__leidenCluster':'cluster'})
```

```python
def enrinch_com(df):
    n_communities = df["cluster"].unique()
    community_dict = {}
    cluster_results = {}


    for cluster in n_communities:
        community_dict[cluster] = df[df['cluster'] == cluster]["name"].to_list()

    for cluster_id, genes in community_dict.items():
        cluster_results[cluster_id] = gprofiler(genes)
    cluster_results = dict(sorted(cluster_results.items()))

    term_filter=["GO:BP","GO:CC","REAC"]
    for cluster_id, cluster_data in cluster_results.items():

        print("-"*60 + f"Cluster {cluster_id}" + "-"*60)

        cluster_data = cluster_data[cluster_data["source"].isin(term_filter)]
        cluster_data=cluster_data[["source","native","name","p_value","description"]]
        cluster_data = cluster_data.sort_values(by=["source", "p_value"])

        cluster_data = cluster_data.groupby("source").head(4)

        cluster_data.reset_index(drop=True, inplace=True)

        
        display(cluster_data)
    
```

```python
enrinch_com(glay)
```

```python
enrinch_com(leinden)
```
