```python
import pandas as pd

def remove_duplicates_with_y_chromosome(df):
    # Group by 'Gene' column and count occurrences
    gene_counts = df.groupby('Gene').size()

    # Find duplicate genes
    duplicate_genes = gene_counts[gene_counts > 1].index

    # Iterate over duplicate genes
    for gene in duplicate_genes:
        # Get indices of rows with the duplicate gene
        duplicate_indices = df[df['Gene'] == gene].index

        # Check if any duplicate row has 'Y' chromosome
        if any(df.loc[duplicate_indices, 'Chromosome'] == 'Y'):
            # Drop rows with 'Y' chromosome
            df = df.drop(duplicate_indices[df.loc[duplicate_indices, 'Chromosome'] == 'Y'])

    return df

# Read the CSV file into a DataFrame
gene_list = pd.read_csv('Results/test_clean.csv')

# Apply the function to remove duplicates with 'Y' chromosome
filtered_gene_list = remove_duplicates_with_y_chromosome(gene_list)


# Sort the filtered DataFrame alphabetically by the 'Gene' column
filtered_gene_list = filtered_gene_list.sort_values(by='Gene')

filtered_gene_list=filtered_gene_list.drop_duplicates(subset='Gene', keep='first').reset_index(drop=True)


gene_counts=filtered_gene_list.groupby('Gene').size()

last_check = filtered_gene_list[filtered_gene_list['Gene'].isin(gene_counts[gene_counts > 1].index)]

filtered_gene_list=filtered_gene_list.sort_values(by="Probability_Class_1",ascending=False).reset_index(drop=True)
# Show the filtered DataFrame
last_check
```

```python
def remove_duplicates_with_y_chromosome(df):
    # Find duplicate genes with more than one occurrence
    duplicate_genes = df[df.duplicated(subset='Gene', keep=False)]

    # Remove rows with 'Y' chromosome for duplicate genes
    df = df[~((df['Gene'].isin(duplicate_genes['Gene'])) & (df['Chromosome'] == 'Y'))]

    return df
```
