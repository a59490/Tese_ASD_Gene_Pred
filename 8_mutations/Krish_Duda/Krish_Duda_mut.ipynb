```python
import pandas as pd



def read_excel_sheets(file_path, sheet_names):
    data = {}
    for sheet_name in sheet_names:
        data[sheet_name] = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=1)
    return data

data = read_excel_sheets('41593_2016_BFnn4353_MOESM13_ESM(2).xlsx', ['prb_rec_LGD', 'prb_LGD', 'prb_SYN', 'sib_LGD', 'sib_SYN'])
```

# Convert Symbol to Ensembl id

```python
from gprofiler import GProfiler

final_gmt = []


for key in data.keys():
    
    NCBI_id = data[key]['gene id'].tolist()
    print(len(NCBI_id))

    gp = GProfiler(return_dataframe=True)
    gp_convert_df =gp.convert(organism  = 'hsapiens',
                                query= NCBI_id,
                                    target_namespace= 'ENSG',
                                    numeric_namespace='ENTREZGENE_ACC',)
    gp_convert_df = gp_convert_df[['incoming', 'converted']]

    ensemble_list = gp_convert_df['converted'].tolist()
    gene_string = '\t'.join(ensemble_list)

    gmt_file= f'{key}\tMutation: {key}\t' + gene_string
    
    final_gmt.append(gmt_file)



```

```python
final_gmt[0]
```

```python
with open('krish_duda.gmt', 'w') as f:
    for line in final_gmt:
        f.write(line + "\n")      
```
